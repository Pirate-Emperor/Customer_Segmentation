```python
data = data.drop(['vendor_discount_amount','delivery_time','order_accepted_time','promo_code','driver_rating','driver_accepted_time','ready_for_pickup_time','picked_up_time','delivered_time','promo_code_discount_percentage','LOCATION_NUMBER','LOCATION_TYPE', 'CID X LOC_NUM X VENDOR'],axis=1)
data.columns
```

### Renaming Columns:


```python
data.rename(columns = {'item_count':'head_count', 'deliverydistance':'vendor_distance', 'preparationtime':'waiting_time','delivery_date':'booked_date',}, inplace = True)
```

```python
data[data.head_count.isnull()].loc[:,['akeed_order_id' ,'head_count']].head()
```

```python
data.drop(data.index[data.akeed_order_id.isnull()], inplace= True)
data.drop(data.index[data.booked_date.isnull()], inplace= True)
```

```python
### Let's also check if the booked_date is greater than the created_at ###
data['booked_date'],data['created_at'] = np.where(data['booked_date'] >= data['created_at'],(data['booked_date'],data['created_at']),(data['created_at'],data['booked_date'])) 
```

### Handling missing values:

```python
data.isnull().sum(axis = 0)   #### Lots of missing values in # of order in last 7 days and last 4 weeks.
```

```python
### Replacing 'Nan' values in head_count in last 7 days and last 4 weeks with 0 ###
data['head_count'][data['head_count'].isnull()] = 1
data['waiting_time'][data['waiting_time'].isnull()] = 0
data['vendor_rating'][data['vendor_rating'].isnull()] = 0
data['is_favorite'][data['is_favorite'].isnull()] = 'No'
```

#### Pareto Analysis (80-20 Rule):

```python
thre_grand_total = 0.8 * np.sum(data['grand_total'])
cum_amt = np.cumsum(data['grand_total'].sort_values(ascending = False))

print('The number of customers responsible for 80% of total grand_total: '+ str(sum(cum_amt <= thre_grand_total)))
print('The percentage of customers responsible for 80% of total grand_total: '+ str(sum(cum_amt <= thre_grand_total) * 100 /data.shape[0]))
```

```python
### Deciding on a date for referencing the current date ###
print('The most recent date overall: '+ str(data['created_at'].max()))

### It is safe to assume that the current date is the most recent date overall. ###
```

### Creating recency Vector

```python
data.reset_index(inplace= True, drop= True)
```

```python
from datetime import datetime
data['recency'] = -(pd.to_datetime(data['created_at']).sub(pd.to_datetime(data['created_at'].max()))).dt.days
data.head(3)
```

### Visualizing data for most valuable customers:

** Changing Datatype of Column ** 

```python
data.columns
data.astype({'head_count':float})
data.head_count.min()
```

```python
### Creating a new table for segmentation and visualizing ###
data_metric = data.loc[:, ['customer_id', 'head_count', 'grand_total', 'recency', 'vendor_id']]

fig, ax = plt.subplots()
graph = ax.scatter(data_metric['head_count'], data_metric['grand_total'], c= data_metric['recency'], cmap='coolwarm', s=10)
cb = fig.colorbar(graph)
cb.set_label('recency', rotation=-90, va='bottom')
ax.margins(0.1)
plt.xlabel('head_count')
plt.ylabel('grand_total')
plt.show()

### To be done later
### Maybe log transformation helps in better visualization ###
fig, ax = plt.subplots()
graph = ax.scatter(np.log(data_metric['head_count']), np.log(data_metric['grand_total']), c= np.log(data_metric['recency']), cmap='coolwarm', s=10)
cb = fig.colorbar(graph)
cb.set_label('Log-recency', rotation=-90, va='bottom')
ax.margins(0.1)
plt.xlabel('Log-head_count')
plt.ylabel('Log-grand_total')
plt.show()
```

```python
print('The quantiles to be used to segment the customers are as follows:')
data_metric.iloc[:, [1, 2, 3]].quantile(q=[0.25,0.5,0.75])
```

```python
data_metric['q_recency'] = pd.qcut(data_metric['recency'], q = 4, labels= [1, 2, 3, 4])
data_metric['q_monetary'] = pd.qcut(data_metric['grand_total'], q = 4, labels= [1, 2, 3, 4])
data_metric['q_frequency'] = pd.qcut(data_metric['head_count'], q = 4, labels= [1, 2, 3], duplicates= 'drop')
```

** Now that the three metrics have been created, we can easilty segment the customers based on this metric. The customers who are most valuable to the company are the ones who have least 'recency_q' and most 'monetary_q' and 'frequency_q'. _i.e. 1-4-3 (R-M-F) customers._ **

```python
### Let's see how our metrics data looks now ###
data_metric.head()
display(data_metric.shape[0])
display(data_metric.customer_id.unique().shape[0])
```

```python
### Finding the most valuable customers ###
val_cust_data = data_metric[(data_metric.q_recency == 1) & (data_metric.q_monetary == 4) & (data_metric.q_frequency == 3)]
val_cust_data.head()
```

```python
### Plotting graph for valuable customers ###
fig, ax = plt.subplots()
graph = ax.scatter(np.log(val_cust_data['head_count']), np.log(val_cust_data['grand_total']), c= np.log(val_cust_data['recency']), cmap='coolwarm', s=10)
cb = fig.colorbar(graph)
cb.set_label('Log-recency', rotation=-90, va='bottom')
ax.margins(0.1)
plt.xlabel('Log-head_count')
plt.ylabel('Log-grand_total')
plt.show()
```

## K-Means clustering for customer segmentation

```python
### Scaling the variables ###
scaler_metric = StandardScaler()
data_metric.iloc[:, [1, 2, 3]] = scaler_metric.fit_transform(data_metric.iloc[:, [1, 2, 3]])
```

```python
data_metric.head(5)   #### Checking ###
```

### Choosing number of clusters:

```python
### Elbow Analysis to find number of clusters ###
n_clusters = range(1, 11)
cluster_sse = []
for i in n_clusters:
    kmeans = KMeans(n_clusters= i)
    kmeans.fit(data_metric.iloc[:, [1, 2, 3]])
    cluster_sse.append(kmeans.inertia_)
    
plt.plot(n_clusters, cluster_sse, marker = "o")
plt.show()
```

### Segmentation into clusters and visualization:

```python
### Segmentation ###
km_cluster = KMeans(n_clusters = 4, random_state= 22)
data_metric['cluster'] = km_cluster.fit_predict(data_metric.iloc[:, [1, 2, 3]])

data_metric.head()
```

```python
data_metric.to_csv('cluster_output.csv')
```

```python
### Let's plot the data points according to their respective clusters ###
fig = plt.figure(figsize= (10, 6))
ax = fig.add_subplot(111, projection='3d')
graph = ax.scatter(xs = data_metric['head_count'], ys = data_metric['grand_total'], zs = data_metric['recency'], c= (data_metric['cluster'] + 1),zdir='z', s=10, depthshade=True)
ax.set_xlabel('head_count')
ax.set_ylabel('grand_total')
ax.set_zlabel('recency')
cb = fig.colorbar(graph, pad=0.1)
cb.set_label('Cluster', rotation= -90, va='bottom')
plt.show()
```

```python
### PCA for better visuaization ###
pca = PCA(n_components=2, whiten=True)
pca.fit(data_metric.iloc[:, [1, 2, 3]])
pca_x = pca.fit_transform(data_metric.iloc[:, [1, 2, 3]])[:, 0]
pca_y = pca.fit_transform(data_metric.iloc[:, [1, 2, 3]])[:, 1]

fig, ax = plt.subplots()
graph = ax.scatter(pca_x, pca_y, c= (data_metric['cluster'] + 1), cmap='viridis', s=10)
cb = fig.colorbar(graph, pad = 0.1)
cb.set_label('Cluster', rotation=-90, va='bottom')
ax.margins(0.1)
plt.xlabel('PCA_X')
plt.ylabel('PCA_Y')
plt.show()
```

## Aggregate variable information by cluster groups

```python
### Information about variables grouped by clusters ###
data['Cluster'] = (data_metric['cluster'] + 1)   ### to account for cluster indices, as index starts from 0 in Python
data['Cluster'].unique()
```

### Obervation
- Cluster 2 has the most recency value which indicates that customer in this cluster were frequent in visiting the restaurant
- First, as suspected, cluster number 4 really outperforms all other clusters ad the customers in that cluster is truly our most valued customer. The company should try to protect them and keep been intact.
- On the other hand, cluster number 3 is the least performing cluster. The customers in this cluster have not ordered since a long time and they tend to not spend a lot too. The company can reach out to these customers for their feedback.
- The selected key performance indicators, 'head_count' (Frequency), 'grand_total' (Monetary) and 'recency' are actually statistically different in each cluster and remain key to segmenting customers into different clusters.
